```python
def upper(x):
    st = ''
    if (ord(x[0]) >= 65) & (ord(x[0]) <= 90):
        st += x[0]
        st += x[1 : len(x)]
    elif (ord(x[0]) >= 97) & (ord(x[0]) <= 122):
        ans = chr(ord(x[0]) - 32)
        st += ans
        st += x[1 : len(x)]
    else:
        return
    return st
```

```python
df['language'] = df['language'].apply(upper)
```

```python
def extraction(x):
    if x == 'Not rated yet':
        return x
    elif (len(x[0]) == 1) & (ord(x[1]) == 32):
        return float(x[0])
    elif (len(x[0 : 3]) == 3) & (ord(x[3]) == 32):
        return float(x[0 : 3])
    else:
        return
```
